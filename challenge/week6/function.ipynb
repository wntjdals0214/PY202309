```python
import random

def get_number():
    return random.randrange(1,46) #랜덥한 수 리턴

def set_lotto_numbers():
    lotto=[]
    while True:
        num = get_number() # 함수 실행
        
        if lotto.count(num) == 0:
            lotto.append(num) #리스트에 랜덤한 수 추가
        
        if len(lotto) >= 6:
            break # 리스트 길이가 6이 되면 중지
    return lotto

def match_numbers(user_inputs, lotto_numbers): # 각각의 리스트 안의 숫자를 하나씩 비교하며 같은 수 개수 찾기
    same_numbers=0
    for num in user_inputs:
        for Num in lotto_numbers:
            if num==Num:
                same_numbers+=1
    return same_numbers

def check_prize(same_numbers): # 같은 수 개수에 따라 등수 매기기
    if same_numbers == 6:
        print("MAtch %d numbers, 1st!!!"%same_numbers)
    elif same_numbers == 5:
        print("MAtch %d numbers, 2nd!!"%same_numbers)
    elif same_numbers == 4:
        print("MAtch %d numbers, 3rd!"%same_numbers)
    elif same_numbers == 3:
        print("MAtch %d numbers, 4th"%same_numbers)
    else:
        print("MAtch %d numbers, no prize :( "%same_numbers)
    
num=0
print("**Start Lotto**")
print("1) Set the Lotto numbers")
lotto_numbers = set_lotto_numbers()

print("2) Set the user inputs")
user_inputs = []
while True:
    num=int(input("ENter %d-number: "%(len(user_inputs)+1)))
    if num < 1 or num> 46: # 입력한 수가 범위를 넘을 경우 다시
        print("No valid number")
        continue
    if user_inputs.count(num)==0:
        user_inputs.append(num)
    else:
        print("Same number") #입력한 수가 중복일 경우 다시
        continue
    if len(user_inputs) >= 6:
        break # 리스트 길이가 6이 되면 중지
        
print("3) Match the numbers")
same_numbers = match_numbers(user_inputs, lotto_numbers)

print("4) Check the prize")
check_prize(same_numbers)

print("5) Sort the lists")
lotto_numbers.sort()
user_inputs.sort() # 리스트 정렬

print("=== Summary ===")
print("Lotto numbers: ", *lotto_numbers)
print("User numbers: ", *user_inputs) # 리스트 출력
        
```
